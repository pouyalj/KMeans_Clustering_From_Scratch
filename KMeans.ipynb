```python
import numpy as np
import random
from scipy.spatial.distance import cdist 
import matplotlib.pyplot as plt
```

```python
pointsM = np.random.rand(2000, 4)
pointsM = pointsM*100
pointsM
```

```python
def kmeans(X, k, max_iter):
    # random choice of indices in X for assigning random centroids
    index = np.random.choice(len(X), k, replace=False)
    centroids = X[index, :]
    # Distances between each centroid and all points
    distances = cdist(X, centroids, 'euclidean')
    # Choosing centroid with min distances to points
    points = np.array([np.argmin(item) for item in distances])
    
    # Loop to repreat the steps above for the maximum iterations equal to max_iter
    for i in range(max_iter):
        centroids = []
        for index in range(k):
            # Updating the centroid by mean of each cluster
            centroids.append(X[points==index].mean(axis=0))
            
        centroids = np.vstack(centroids)
        # Same as above, Distances between each centroid and all points
        distances = cdist(X, centroids, 'euclidean')
        # Same as above, Choosing centroid with min distances to points
        points = np.array([np.argmin(item) for item in distances])
    return points
```

```python
groups = kmeans(pointsM, 5, 1000)
Ugroups = np.unique(groups)
```

```python
listed_by_groups = {}
for G in Ugroups:
    listed_by_groups[G] = pointsM[groups == G]
listed_by_groups
```

### For 2D (2 Features) with Ilustration

```python
P2D = np.random.rand(2000, 2)
P2D = P2D*100
```

```python
groups = kmeans(P2D, 7, 1000)
Ugroups = np.unique(groups)
```

```python
listed_by_groups = {}
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(15)
for G in Ugroups:
    listed_by_groups[G] = P2D[groups == G]
    plt.scatter(P2D[groups == G, 0], P2D[groups == G, 1], label = G)

plt.legend()
plt.show()
```
